####ASSUMPTION #1: linearity and additivity of relationship between IVs and DV:


Explanation: 

    The expected value of DV is a linear function of each IV, holding the others fixed (with no interactions), and the effects of different IVs on the expected value of the DV are additive.

Checks:
    
    (a) eyeball scatter plots/linear fits
    (b) plot residuals as a function of x (bow shape suggests that more extreme values of x are providing a worse fit... possibly nonlinear relationship needs to be modeled)

Solutions:

    (a) apply nonlinear transform to IV (e.g. if Y varies linearly with X**2, use the X**2 transformed IV)
    (b) some relationships may be multiplicative rather than additive (e.g. a small percentage change in an IV induces a proportional percentage change in the DV). this can be handled by applying a log transformation to both 
    (c) fit an interaction regressor to linearly model the iteracctions between IVs

####ASSUMPTION #2: homoscedasticity (constant variance) of the errors

Explanation:
    
    variance of the errors should be constant across IVs (including time in the case of time series data) and constant across the range of predicted values.
    
Problem:

    heteroscedasticity makes it difficult to gauge the true standard deviation of the forecast errors, meaning that your confidence intervals can be either too wide or two narrow. for example, if the error variance increases as a function of X, then you may have overly narrow confidence intervals when predicting out of sample (higher) values of X.

Checks:

    (a) plot residuals as a function of the predicted value: heteroscedasticity could be visible as a relationship between variance/spread of the errors and the predicted value (this is the most important one to do)
    (b) plot residuals as a function of x (including time): heteroscedasticity could be visible as a relationship between variance/spread of the errors and the DV
   
Solutions:

    (a) if the residual-versus-predicted plot shows that the size of the errors is proportional to the size of the predictions (i.e. the errors seem constant in percentage rather than absolute terms), you can apply a log transformation to the DV
    (b) fit autoregressive model

```python
def resid_vs_predicted(df, x, y):
    from pandas.stats.api import ols
    model = ols(x=df[x], y=df[y])
    predicted=model.y_fitted
    df=pd.DataFrame(data={'residuals':model.resid, 'predicted %s'%y:model.y_fitted})
    sns.regplot('predicted %s'%y, 'residuals', df)
    
```

####ASSUMPTION #3: statistical independence of the errors 

Explanation:
    
    no auto-correlation (correlation between consecutive errors in the case of time series data). assumption is violated if errors tend to always have the same sign under particular conditions (i.e. if the model systematically underpredicts or overpredicts for certain values of the IV). this kind of violation is particularly common and problematic with time-series data where events nearby in time tend to have systematic errors.  note: this is different from homoscedasticity in that this assumption pertains to the actual magnitude and sign of the errors, whereas homoscedasticity refers to the variance of the errors.
    
Problem:
    
    often means that your model is poorly specified and that you need to model certain periodic effects (e.g. day of the week, seasonality, etc.)
    
Checks:

    (a) plot residuals as a function of x (e.g. time): if there is auto-correlation, there will be a relationship between the direction/size of the error and the IV
    (b) plot of residual autocorrelations
    (c) Durbin-Watson statistic
    
Solutions:

    (a) add lagged regressors
    (b) fit an ARMA model
    (c) take another look at variable transformations

####ASSUMPTION #4: normality of residuals

Explanation:
    
    Errors are normally distributed (no big outliers, etc.). Can occur because the distributions of the DV and/or IVs are themselves significantly non-normal, and/or the linearity assumption is violated
    
Problem:
    
    Variance of errors is used to determine  whether model coefficients are significantly different from zero and for calculating confidence intervals for predictions. outliers/non-normally distributed errors can lead to incorrect confidence intervals/hypothesis testing
    
Checks

    (a) plot histogram of residuals
    (b) qqplot
    (c) KS test for normality of residuals
    
Solutions

    (a) nonlinear transformation of one or more variables can fix non-normality, non-linearity

```python
def resid_normality(df, x, y):
    from pandas.stats.api import ols
    model = ols(x=df[x], y=df[y])
    plt.hist(model.resid, bins=20)
    f,ax=plt.subplots(figsize=[4,4])
    scipy.stats.probplot(model.resid, dist="norm", plot=plt)
    ax.set_title('qqplot')
    ksstat, pvalue= scipy.stats.kstest(model.resid, 'norm')
    print "ks statistic=%.3f, p=%.3f" %(ksstat, pvalue)
resid_normality(df, 'B', 'Y')
```

```python
%%writefile scripts/linear_regression.py

import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

def clean_data(loansData):
    loansData['Interest.Rate']=loansData['Interest.Rate'].apply(lambda x: float(x[:-1]))
    loansData['Loan.Length']=loansData['Loan.Length'].apply(lambda x: float(x[:-7]))
    loansData['FICO.Score']=loansData['FICO.Range'].apply(lambda x: float(x.split('-')[0]))
    return loansData

def exploratory_viz(loansData):
    plt.figure()
    p = loansData['FICO.Score'].hist()
    plt.savefig('../figs/fico_score_hist.png')

    a = pd.scatter_matrix(loansData, alpha=0.05, figsize=(14,14))
    plt.savefig('../figs/loan_scatter_matrix.png')

def extra_viz(loansData):
    f, ax = plt.subplots(figsize=(10, 10))
    sns.corrplot(loansData, ax=ax)
    plt.savefig('../figs/loan_corr_matrix.png')
    
def model(loansData, iv1, iv2, dv):    
    intrate = loansData[dv]
    fico = loansData[iv1]
    loanamt = loansData[iv2]

    # The dependent variable
    y = np.matrix(intrate).transpose()
    # The independent variables shaped as columns
    x1 = np.matrix(fico).transpose()
    x2 = np.matrix(loanamt).transpose()

    x = np.column_stack([x1,x2])

    X = sm.add_constant(x)
    model = sm.OLS(y,X)
    return model.fit()

def model2(loansData, iv1, iv2, dv):    
    for var in [iv1, iv2, dv]:
        loansData=loansData.rename(columns={var:''.join([char for char in var if char !='.'])})
    iv1=''.join([char for char in iv1 if char !='.'])
    iv2=''.join([char for char in iv2 if char !='.'])
    dv=''.join([char for char in dv if char !='.'])
    model = sm.OLS.from_formula('%s ~ %s + %s' %(dv, iv1, iv2), loansData)
    return model.fit()


if __name__=="__main__":
    loansData = pd.read_csv('https://spark-public.s3.amazonaws.com/dataanalysis/loansData.csv')
    loansData=clean_data(loansData)
    exploratory_viz(loansData)
    extra_viz(loansData)
    f=model2(loansData, 'FICO.Score', 'Amount.Requested', 'Interest.Rate')
    print f.summary()
```

```python
def coefplot(dv, iv1, iv2, loansData):
    for var in [iv1, iv2, dv]:
        loansData=loansData.rename(columns={var:''.join([char for char in var if char !='.'])})
    iv1=''.join([char for char in iv1 if char !='.'])
    iv2=''.join([char for char in iv2 if char !='.'])
    dv=''.join([char for char in dv if char !='.'])
    fml='%s ~ %s + %s' %(dv, iv1, iv2)
    print fml
    sns.coefplot(fml, loansData)
coefplot('Interest.Rate', 'FICO.Score', 'Amount.Requested', loansData)
```
